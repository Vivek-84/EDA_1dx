```python
## For Normality(MinMaxScaler)
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
min_max.fit_transform(dff[['fixed acidity', 'residual sugar',
       'free sulfur dioxide', 'total sulfur dioxide',
       'pH', 'alcohol', 'quality']])
```

# Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?



```python
dff1.head()
```

```python
#Before applying PCA or any other Machine Learning technique it is always considered good practice to standardize the data.
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
pca.fit(dff1)
data_pca = pca.transform(dff1)
data_pca = pd.DataFrame(data_pca,columns=['PC1','PC2','PC3'])
data_pca.head()
```

```python
# The plot representing the variance explained by the components is called Scree Plot. The first 2 components explain 
# approximately 80% of the total variation and 3 components, almost 90%.
```

```python
# Now that we have applied PCA and obtained the reduced feature set, we will check the co-relation between various Principal 
# Components, again by using a heatmap.
```

```python
#Checking Co-relation between features after PCA
sns.heatmap(data_pca.corr())
```
